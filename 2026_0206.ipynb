```python
import numpy as np
import math

def gauss_1d_c():
  # Basis of 1D Gaussians, with varying centers,
  # all with same width

  n = 101
  
  # Calculate the centers for the Gaussians
  s = np.zeros(n)
  v = np.zeros(n)
  for i in range(n):
    s[i] = -25.0 + (i - 1) * 0.5
    v[i] = 0.5 + (i-1)*0.01          
  # Setup the Hamiltonian
  h = np.zeros((n, n))
  o = np.zeros((n, n))
  f = np.zeros((n, n))
  for i in range(n):
    for j in range(n):
      ss = (s[i] - s[j])**2
      o[i, j] = math.sqrt(math.pi/(v[i]+v[j]))*np.exp(-( (v[i]*v[j])/(v[i]+v[j]) ) * ss)  #Gaussiams Function
      p = 0.5*(1/(2*(v[i]+v[j])) + ((s[i]*v[i]+s[j]*v[j])/((v[i]+v[j])))**2 )*o[i, j]
      t = ( (v[i]*v[j])/(v[j]+v[i]) ) *(1 - ss*2*(v[i]*v[j])/(v[j]+v[i]) )*o[i, j]
      h[i, j] = t + p

  # Diagonalize
  f=np.dot(h,np.linalg.inv(o))
  eigenvalues, eigenvectors = np.linalg.eig(f)
  print(eigenvalues.min())
  
gauss_1d_c()
```

在这里，仅考虑一维谐振子势场  
V(x) = (1/2)mω²x²  
哈密顿量为：  
H = -ħ²/(2m) d²/dx² + (1/2)mω²x²  

选择高斯型基函数  
$\varphi_i(x) = \exp(-\nu_i (x - s_i)^2)$，其中 $\nu_i > 0$，$s_i$ 是中心点。
两个高斯函数的内积为：  
$O_{ij}$ = ⟨φᵢ|φⱼ⟩ = ∫ φᵢ(x) φⱼ(x) dx  
$O_{ij} = \sqrt{\frac{\pi}{\nu_i+\nu_j}} \exp\left(-\frac{\nu_i*\nu_j}{\nu_i+\nu_j}  (s_i - s_j)^2\right)$  
总哈密顿矩阵为  
$H_{ij}=P_{ij}+T_{ij}$  
其中，势能矩阵元为$P_{ij}$ = ⟨φᵢ | 1/2 ωx² | φⱼ⟩,  动能矩阵元为$T_{ij}$ =⟨φᵢ | -ħ²/(2m) d²/dx² | φⱼ⟩  
为简化计算，取自然单位制：ħ = 1, m = 1, ω = 1，v=1，积分得到：  
$P_{ij} = O_{ij} \cdot \frac{1}{2} \left[ \frac{1}{2*(\nu_j+\nu_i)} + (\frac{\nu_i*s_i+\nu_j*s_j}{2*(\nu_j+\nu_i)})^2 \right]$  
$T_{ij} = O_{ij} \cdot \frac{\nu_i*\nu_j}{\nu_i+\nu_j} \left[ 1 - 2*\frac{\nu_i*\nu_j}{\nu_i+\nu_j}(s_i - s_j)^2 \right]$  
$s_i,s_j，v_i,v_j$为可调参数  
问题变成：  
$\sum_{i=1}^{N} H_{ij} c_{i} = \epsilon_k \sum_{i=1}^{N} O_{ij} c_{i}$
  
通过迭代
$s_i,s_j,v_i,v_j得到最小的\epsilon_k，此时为本征能量，此时s_i,s_j下对应的波函数为本征波函数$

```python
import numpy as np
import math

def gauss_1d_c():
  # Basis of 1D Gaussians, with varying centers,
  # all with same width

  n = 201
  v = 1.0 
  # Calculate the centers for the Gaussians
  s = np.zeros(n)
  for i in range(n):
    s[i] = -50.0 + (i - 1) * 0.5

  # Setup the Hamiltonian
  h = np.zeros((n, n))
  o = np.zeros((n, n))
  f = np.zeros((n, n))
  for i in range(n):
    for j in range(n):
      ss1 = (s[i] + s[j])**2
      ss2 = (s[i] - s[j])**2
      o[i, j] = math.sqrt(math.pi/(2*v))*np.exp(-0.5* v * ss2)  #Gaussiams Function
      p = 0.5*(0.25 + 0.25*ss1 )*o[i, j]
      t = 0.5*(1 - ss2 )*o[i, j]
      h[i, j] = t + p

  # Diagonalize
  f=np.dot(h,np.linalg.inv(o))
  eigenvalues, eigenvectors = np.linalg.eig(f)
  print(eigenvalues.min())
  
gauss_1d_c()
```
